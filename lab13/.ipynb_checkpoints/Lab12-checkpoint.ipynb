## Problem 1 : Principal Component Analysis, PCA (20 points)

Given a dataset, PCA tries to find principal axes of the dataset, those that well describe the dataset. PCA returns the axis along which the data has the highest variance. To understand PCA, let's look at the following example.

We are given a dataset consisting of $200$ points distributed in 2D space. Below we define and plot this dataset:

```python
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T

plt.figure()
plt.scatter(X[:, 0], X[:, 1], color='blue')
plt.title('Vizualization of the dataset')
plt.axis('equal');
```

We observe a relationship between the $x$ and $y$ axes of the points in the dataset (it seems like there is a linear relationship between the $x$ and $y$ coordinates of this dataset). Given the above dataset let's see the principal axis output by PCA. For this purpose, we use sklearn library.

```python
pca = PCA(n_components=2)
pca.fit(X)

princ_axes = pca.components_
princ_std = np.sqrt(pca.explained_variance_) * 3.
```

Now let's visualize the principal axes alongside the data.

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1], color='blue')

plt.plot(np.array([0, princ_axes[0, 0] * princ_std[0]]), 
         np.array([0, princ_axes[1, 0] * princ_std[0]]), 
         color='red',
         label='Principal axis #1')
plt.plot(np.array([0, princ_axes[1, 0] * princ_std[1]]), 
         np.array([0, princ_axes[1, 1] * princ_std[1]]), 
         color='orange', 
         label='Principal axis #2')

plt.title('Vizualization of the dataset + principal axes')
plt.legend()
plt.axis('equal');
```

The above visualization shows that principal axes. The length of the axes is related to the variance of the dataset along that axis. Therefore, the dataset has a larger variance along the first principal axis compared to the second principal axis. 

This means we can choose to represent the dataset based on the first principal axis. In other words, we can project the dataset on the first principal axis and discard the remaining part of the data projected on the second principal axis. Let's check this visually.

```python
# project the dataset onto the first principal axis
pca = PCA(n_components = 1).fit(X)
components = pca.transform(X)
filtered = pca.inverse_transform(components)
```

```python
plt.figure()
plt.scatter(X[0, 0], X[0, 1], alpha=0.5, color='blue', label='Original data')
plt.scatter(filtered[0, 0], filtered[0, 1], color='orange', label='Projected data on 1st princ. axis')
plt.scatter(X[:, 0], X[:, 1], alpha=0.5, color='blue')
plt.scatter(filtered[:, 0], filtered[:, 1], color='orange')
plt.legend()
plt.show()
```

Note that after the projection on the first principal axis, the projected data points lie on the first principal axis.

### Problem 1-1: PCA for dimensionality reduction (10 points)

One application of PCA is reducing the dimentionality of a high-dimensional dataset. This is helpful when we want to visualize a high dimensional dataset and its underlying structure. Let's assume we have a dataset ${X}\in \mathbb{R}^{L \times n}$. This dataset contains $n$ samples. Each of the samples are a vector of length $L$, where $L$ can be larger than $3$. For $L>3$ we cannot visualize this dataset in its original form. Thus, we need to reduce its dimensionality (by projecting the dataset onto a subset of the prinicipal axis).

Let's list the PCA steps and how we can project the dataset on a $k=2$ of the principal axes.
* Construct the mean of the samples. Let's call this mean $\overline{X}$. It should be a vector of length $L$.
* Construct the covarinace matrix as:
\begin{align}
C = \frac{1}{n} (X-\overline{X}) (X-\overline{X})^T 
\tag 1
\end{align}
* Perform eigenvalue decomposition on C. This would give you a set of eigenvectors and eigenvalues of $C$.
* Choose $K$ top eigenvectors, i.e. the eigenvectors associated with the $k$ largest eigenvalues. These vectores will be your $k$ prinicipal axes. Lets call these $k$ vectors $W \in \mathbb{R}^{L \times k}$
* Find the components of the data samples along the principal axes. For any data sample $x \in X$, this is obtained by:
\begin{align}
\widehat{x} = [<x, u_1>, <x, u_2>, ..., <x, u_K>]
\tag 2
\end{align}
where $\widehat{x}$ is the projected sample on the top $K$ eigenvectors and $\{u_k\}_{k=1}^K$ denote the top-$K$ eigenvectors.
***

Based on the above descriptions, in this problem, we aim to use PCA to reduce the dimension of a dataset and visulize the projected dataset and its structure.

Inputs:
* The dataset $\mathcal{D} \in \mathbb{R}^{L \times n}$, a np.ndarray of shape $L \times n$ where $L$ is the dimension of the data samples (i.e. each data sample is a vector of length $L$) and $n$ is the number of samples.

Outputs:
* A np.ndarray of shape $2 \times n$. The output corresponds to the projection of the input to the first two principal components.

```python
def pca_2(X_train):
    num_samples = X_train.shape[1]
    x_mean = np.mean(X_train, axis=1)
    tmp = X_train - np.expand_dims(x_mean, 1)
    C = np.dot(tmp, tmp.T) / num_samples

    k = 2
    w, v = np.linalg.eigh(C)
    index = np.argsort(w)
    v = v[:, index[::-1]]

    # projecting data
    out = np.dot(v[:, 0:k].T, X_train)
    return out
```

### Problem 1-2: PCA for denoising (10 points)
Another application of PCA is denoising. Looking back at the example provided at the begining of Problem 1, after projecting the data on the first prinicipal component we sort of denoised the dataset!

In this problem, our goal is to project a dataset containing noisy samples on a subset of the principal axis. This helps in denoising the dataset. To obtain the projection of the dataset on the top $K$ principal axes , you need to follow the same steps as discussed in Problem 1-1, except the last step. You need to replace the last step with:
* 
\begin{align}
\widehat{x} = \sum\limits_{k=1}^K <x, u_k> u_k
\tag 3
\end{align}
where $x \in X$, $\widehat{x}$ denotes $x$ projected onto the $K$ principal axes.
***
Write a function that denoises a set of noisy samples by projecting the dataset onto $K$ principal axes. 

The inputs:
* The dataset $X$, a np.ndarray matrix of size $L \times n$. Here, $L$ denotes the dimentionality of the samples and $n$ is the number of samples.
* $K$, a scalar dentoting the number of principal components used in PCA.

The outputs:
* A np.ndarray matrix of size $L \times n$ containing the denoised samples.

Add your code to `Question1.pca_project` in `main.py`. You are <b>NOT</b> allowed to use `sklearn` library and are asked to implement PCA from scratch. You can compare your PCA implementation with the results from `sklearn` functions.

```python
def pca_project(X_train, k):
    num_samples = X_train.shape[1]
    x_mean = np.mean(X_train, axis=1)
    tmp = X_train - np.expand_dims(x_mean, 1)
    C = np.dot(tmp, tmp.T) / num_samples

    w, v = np.linalg.eigh(C)
    index = np.argsort(w)
    v = v[:, index[::-1]]

    # projecting data
    out = np.dot(v[:, 0:k].T, X_train)
    filtered = np.dot(v[:, 0:k], out)
    print(out.shape)
    return filtered
```

```python
def pca_project_sklearn(X_train, k):
    pca = PCA(n_components = k).fit(X_train)
    components = pca.transform(X_train)
    filtered = pca.inverse_transform(components)
    return filtered
```

### Visualization: Embedding of a ring graph

```python
def construct_ring_graph(mat_sz):
    # ring graph
    row1 = np.zeros((mat_sz,))
    row1[1] = 0.5
    row1[-1] = 0.5
    W = scipy.linalg.circulant(row1)
    print(W)
    return W
```

```python
def construct_complete_graph(mat_sz):
    W = np.ones((mat_sz, mat_sz)) - np.eye(mat_sz)
    W /= (mat_sz-1)
    print(W)
    return W
```

```python
n = 12
W = construct_ring_graph(n)

D = np.diag(np.sum(W, 1)**(-0.5))
W_prime = np.dot(D, np.dot(W, D))
# w, v = np.linalg.eigh(W_prime)
v, w = q3.embedding(W_prime)
# print(v)

fig = plt.figure()
ax = fig.add_subplot(111)
plt.scatter(v[:, 0], v[:, 1])
plt.title('Embedding plot of ring graph')
ax.set_aspect('equal')
plt.show()

# eigen-value plots
plt.figure()
plt.plot(w, marker='x')
plt.ylabel('Eigen values')
plt.show()
```
